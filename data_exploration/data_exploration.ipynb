```python
import pandas as pd
import joblib as jb
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator, FuncFormatter
import numpy as np

plt.rcdefaults()

# Package versions
print(f"Pandas version: {pd.__version__}")
print(f"Joblib version: {jb.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print(f"Numpy version: {np.__version__}")
```

```python
# Loading data
data_per_year = jb.load("data_per_year.pkl")
```

```python
# Printing all columns from the dataset
for col in data_per_year[2015].columns:
    print(col)
```

```python
# Total instances and % of no-shows for each year

# Extract years, scheduled counts, no-show counts, and compute no-show rate
years = list(data_per_year.keys())
scheduled = [len(data_per_year[y]) for y in years]
no_shows = [len(data_per_year[y][data_per_year[y]["no_show"] == 1]) for y in years]
no_show_rate = [ns / tot * 100 if tot > 0 else 0 for ns, tot in zip(no_shows, scheduled)]

# Create figure and first axis
fig, ax1 = plt.subplots(figsize=(9, 4))

# Bar plots on ax1
ax1.bar(years, scheduled, alpha=0.5,
        label="Scheduled", edgecolor="black", color="gray")
ax1.bar(years, no_shows, alpha=0.7,
        label="No-show",   edgecolor="black", color="black")

# Annotate each bar with "scheduled/no-show"
for y, tot, ns in zip(years, scheduled, no_shows):
    ax1.annotate(f"{tot}/{ns}", xy=(y, tot), ha="center", va="bottom")

# Format ax1
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of instances")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

# Create secondary axis for no-show rate
ax2 = ax1.twinx()
ax2.plot(years, no_show_rate,
         marker="o", linestyle=":",
         label="No-show (%)", color="black")

# Switch ax2 to log scale and set ticks for percentages
# (exclude zeros by starting y-limit at the smallest positive rate)
min_pos = min(r for r in no_show_rate if r > 0)
ax2.set_yscale("log")
ax2.yaxis.set_major_locator(LogLocator(base=10.0,
                                       subs=range(1, 10),
                                       numticks=15))
ax2.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.0f}%"))
ax2.set_ylim(min_pos * 0.8, max(no_show_rate) * 1.1)

# Format ax2
ax2.set_ylabel("No-show (%)")
ax2.tick_params(axis='y', labelsize=8)

# Combine legends
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc="upper left")

plt.xticks(list(data_per_year.keys()))
plt.gca().spines['top'].set_visible(False)
plt.tight_layout()
plt.savefig("by_year.png", dpi=400)
plt.show()
```

```python
# Concatenating all the data
full_data = pd.DataFrame()
for year in data_per_year.keys():
    full_data = pd.concat([full_data, data_per_year[year]]).reset_index(drop=True) # Reset index to avoid any possible problem
print(f"Total number of instances: {len(full_data)}")
```

```python
# Note that there is missing data only for the columns "sex", "age", "distance_to_clinic" and "no_show_history".
full_data.isna().sum()[full_data.isna().sum() > 0]
```

```python
# When it is the patient's first time at the clinic, the attendants usually leave the data to be recorded only when the patient goes to the clinic.
# From the point of view of no-show prediction, this generates data leakage, because if there is NaN in the "no_show_history" and the patient does not
# show up, the variables "sex", "age" and "distance_to_clinic" remain unfilled, so these missing values ​​are highly correlated with no-shows.
# Missing values ​​for "no_show_history", on the other hand, only indicate that it is the patient's first time at the clinic.
temp = full_data[["distance_to_clinic", "sex", "age", "no_show_history", "no_show"]].copy()
temp["missing_sex"] = temp["sex"].isna()
temp["missing_age"] = temp["age"].isna()
temp["missing_distance_to_clinic"] = temp["distance_to_clinic"].isna()
temp["missing_no_show_history"] = temp["no_show_history"].isna()

wide = temp.pivot_table(
    index=[
        "missing_distance_to_clinic",
        "missing_sex",
        "missing_age",
        "missing_no_show_history"
    ],
    columns="no_show",
    aggfunc="size",
    fill_value=0
)

long = (
    wide
    .reset_index()
    .melt(
        id_vars=[
            "missing_no_show_history",
            "missing_distance_to_clinic",
            "missing_sex",
            "missing_age"
        ],
        value_vars=[0.0, 1.0],
        var_name="no_show",
        value_name="count"
    )
)

long["proportion"] = (
    long
    .groupby([
        "missing_no_show_history",
        "missing_distance_to_clinic",
        "missing_sex",
        "missing_age"
    ])["count"]
    .transform(lambda x: x / x.sum())
)

long.sort_values(by=["missing_no_show_history", "missing_distance_to_clinic", "missing_sex", "missing_age", "no_show"])
```

```python
# Total number of instances and % no-show by patient sex

# Copy and normalize the 'sex' column
temp = full_data[["sex", "no_show"]].copy()
temp["sex"] = temp["sex"].astype(str)
temp.loc[temp["sex"] == "M",   "sex"] = "Male"
temp.loc[temp["sex"] == "F",   "sex"] = "Female"
temp.loc[temp["sex"] == "nan", "sex"] = "Missing"

# Calculate scheduled counts, no-show counts, and rates
sex_categories = list(temp["sex"].unique())
scheduled      = [len(temp[temp["sex"] == s])                         for s in sex_categories]
no_shows       = [len(temp[(temp["sex"] == s) & (temp["no_show"] == 1)]) for s in sex_categories]
no_show_rate   = [ns / tot * 100 if tot > 0 else 0
                  for ns, tot in zip(no_shows, scheduled)]

# Create figure & primary axis
fig, ax1 = plt.subplots(figsize=(6, 4))

# Plot bars for scheduled and no-show
ax1.bar(sex_categories, scheduled, alpha=0.5,
        label="Scheduled", edgecolor="black", color="gray")
ax1.bar(sex_categories, no_shows, alpha=0.7,
        label="No-show",   edgecolor="black", color="black")

# Annotate each bar with "scheduled/no-show"
for s, tot, ns in zip(sex_categories, scheduled, no_shows):
    ax1.annotate(f"{tot}/{ns}", xy=(s, tot),
                 ha="center", va="bottom")

# Format primary axis
ax1.set_ylabel("Number of instances")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
plt.xticks(sex_categories)

# Create secondary axis for no-show rate
ax2 = ax1.twinx()
ax2.plot(sex_categories, no_show_rate,
         marker="o", linestyle=":",
         label="No-show (%)", color="black")

# Switch secondary axis to log scale and configure ticks
min_pos = min(r for r in no_show_rate if r > 0)
ax2.set_yscale("log")
ax2.yaxis.set_major_locator(LogLocator(base=10.0,
                                       subs=range(1, 10),
                                       numticks=15))
ax2.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.0f}%"))
ax2.set_ylim(min_pos * 0.8, max(no_show_rate) * 1.1)

# Format secondary axis
ax2.set_ylabel("No-show (%)")
ax2.tick_params(axis="y", labelsize=8)

# Combine legends and finalize
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc="upper center")

plt.gca().spines['top'].set_visible(False)
plt.tight_layout()
plt.savefig("by_sex.png", dpi=400)
plt.show()
```

```python
# Total number of instances and % no-show by patient age

# Copy the relevant columns
temp = full_data[["age", "no_show"]].copy()

# Bin ages into 10 quantile-based bins
age_bins = pd.qcut(temp["age"], q=10, duplicates="drop")
temp["age_bin"] = age_bins.astype(str)

# Label missing ages as "Missing"
temp.loc[temp["age"].isna(), "age_bin"] = "Missing"

# Build an ordered list of all bins (quantiles plus "Missing")
quantile_labels = age_bins.cat.categories.astype(str).tolist()
bin_order = quantile_labels + ["Missing"]

# Compute scheduled and no-show counts per bin
scheduled    = [len(temp[temp["age_bin"] == b]) for b in bin_order]
no_shows     = [len(temp[(temp["age_bin"] == b) & (temp["no_show"] == 1)]) for b in bin_order]
# Compute no-show rate in %
no_show_rate = [ns / tot * 100 if tot > 0 else 0 for ns, tot in zip(no_shows, scheduled)]

# Create figure and first axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot bars on ax1
ax1.bar(bin_order, scheduled, alpha=0.5,
        label="Scheduled", edgecolor="black", color="gray")
ax1.bar(bin_order, no_shows, alpha=0.7,
        label="No-show", edgecolor="black", color="black")

# Annotate each bar with "scheduled/no-show"
for b, tot, ns in zip(bin_order, scheduled, no_shows):
    ax1.annotate(f"{tot}/{ns}", xy=(b, tot),
                 ha="center", va="bottom")

ax1.set_xlabel("Age (years) bin")
ax1.set_ylabel("Number of instances")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.set_ylim(0, 4000)

# Create secondary axis for no-show rate
ax2 = ax1.twinx()
ax2.plot(bin_order, no_show_rate,
         marker="o", linestyle=":",
         label="No-show (%)", color="black")

# Switch secondary axis to log scale and configure ticks
min_pos = min(r for r in no_show_rate if r > 0)
ax2.set_yscale("log")
ax2.yaxis.set_major_locator(LogLocator(base=10.0,
                                       subs=range(1, 10),
                                       numticks=15))
ax2.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.0f}%"))
ax2.set_ylim(min_pos * 0.8, max(no_show_rate) * 1.1)
ax2.set_ylabel("No-show (%)")
ax2.tick_params(axis="y", labelsize=8)

# Combine legends and finalize
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc="upper center")

# Rotate x-tick labels and layout
plt.setp(ax1.get_xticklabels(), rotation=45)
plt.gca().spines['top'].set_visible(False)
plt.tight_layout()
plt.savefig("by_age.png", dpi=400)
plt.show()
```

```python
# Total instances and % no-show by month

# Copy and normalize 'month'
temp = full_data[["month", "no_show"]].copy()
month_map = {
    1: "January",   2: "February", 3: "March",
    4: "April",     5: "May",      6: "June",
    7: "July",      8: "August",   9: "September",
    10: "October", 11: "November", 12: "December"
}
temp["month_name"] = temp["month"].map(month_map)

# Define categories in calendar order and calculate counts and rates
month_categories = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]
scheduled    = [len(temp[temp["month_name"] == m]) for m in month_categories]
no_shows     = [
    len(temp[(temp["month_name"] == m) & (temp["no_show"] == 1)])
    for m in month_categories
]
no_show_rate = [
    (ns / tot * 100) if tot > 0 else 0
    for ns, tot in zip(no_shows, scheduled)
]

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(11, 4))

# Plot bars for scheduled and no-show
ax1.bar(month_categories, scheduled, alpha=0.5,
        label="Scheduled", edgecolor="black", color="gray")
ax1.bar(month_categories, no_shows, alpha=0.7,
        label="No-show", edgecolor="black", color="black")
ax1.set_ylim(0, 3500)

# Annotate each bar with "scheduled/no-show"
for month, tot, ns in zip(month_categories, scheduled, no_shows):
    ax1.annotate(f"{tot}/{ns}", xy=(month, tot),
                 ha="center", va="bottom")

# Format primary axis
ax1.set_ylabel("Number of instances")
plt.xticks(rotation=45)

# Create secondary axis for no-show rate
ax2 = ax1.twinx()
ax2.plot(month_categories, no_show_rate,
         marker="o", linestyle=":",
         label="No-show (%)", color="black")

# Set log scale and configure ticks on secondary axis
min_positive = min(r for r in no_show_rate if r > 0)
ax2.set_yscale("log")
ax2.yaxis.set_major_locator(LogLocator(
    base=10.0, subs=range(1, 10), numticks=15
))
ax2.yaxis.set_major_formatter(FuncFormatter(
    lambda y, _: f"{y:.0f}%"
))
ax2.set_ylim(min_positive * 0.8, max(no_show_rate) * 1.5)

# Format secondary axis
ax2.set_ylabel("No-show (%)")
ax2.tick_params(axis="y", labelsize=8)

# Combine legends and finalize layout
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles1 + handles2, labels1 + labels2,
           loc="upper left")
ax1.spines["top"].set_visible(False)
ax2.spines["top"].set_visible(False)
plt.tight_layout()
plt.savefig("by_month.png", dpi=400)
plt.show()
```

```python
# Total instances and % no-show by hour

# Copy 'hour' and 'no_show', generate "HH:00" labels
temp = full_data[["hour", "no_show"]].copy()
temp["hour_label"] = temp["hour"].apply(lambda h: f"{h:02d}:00")

# Keep only clinic hours 6:00–22:00 and define categories
temp = temp[(temp["hour"] >= 6) & (temp["hour"] <= 22)]
hour_categories = [f"{h:02d}:00" for h in range(6, 23)]

# Compute scheduled counts, no-show counts, and rates
scheduled    = [len(temp[temp["hour_label"] == hl]) for hl in hour_categories]
no_shows     = [
    len(temp[(temp["hour_label"] == hl) & (temp["no_show"] == 1)])
    for hl in hour_categories
]
no_show_rate = [
    (ns / tot * 100) if tot > 0 else 0
    for ns, tot in zip(no_shows, scheduled)
]

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.bar(hour_categories, scheduled, alpha=0.5,
        label="Scheduled", edgecolor="black", color="gray")
ax1.bar(hour_categories, no_shows, alpha=0.7,
        label="No-show", edgecolor="black", color="black")

# Annotate each bar with "scheduled/no-show"
for hl, tot, ns in zip(hour_categories, scheduled, no_shows):
    ax1.annotate(f"{tot}/{ns}", xy=(hl, tot+25),
                 ha="center", va="bottom", fontsize=8)

# Format primary axis
ax1.set_xlabel("Hour of Day")
ax1.set_ylabel("Number of instances")
plt.xticks(rotation=90)

# Prepare line plot data for hours 07:00–22:00 only
line_hours = [f"{h:02d}:00" for h in range(7, 23)]
line_rates = [no_show_rate[hour_categories.index(lh)] for lh in line_hours]

# Create secondary axis and plot line for 07:00–22:00
ax2 = ax1.twinx()
ax2.plot(line_hours, line_rates,
         marker="o", linestyle=":", label="No-show (%)", color="black")

# Set log scale and configure ticks on secondary axis
min_positive = min(r for r in line_rates if r > 0)
ax2.set_yscale("log")
ax2.yaxis.set_major_locator(LogLocator(
    base=10.0, subs=range(1, 20), numticks=15
))
ax2.yaxis.set_major_formatter(FuncFormatter(
    lambda y, _: f"{y:.0f}%"
))
ax2.set_ylim(min_positive * 0.8, max(line_rates) * 1.1)

# Format secondary axis and combine legends
ax2.set_ylabel("No-show (%)")
ax2.tick_params(axis="y", labelsize=8)
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2,
           loc="upper center")

ax1.spines["top"].set_visible(False)
ax2.spines["top"].set_visible(False)
plt.tight_layout()
plt.savefig("by_hour.png", dpi=400)
plt.show()
```

```python
# Total instances and % no-show considering holidays for yesterday and tomorrow

# Copy and normalize holiday flags
temp = full_data[['was_holiday_yesterday', 'is_holiday_tomorrow', 'no_show']].copy()
bool_map = {0: 'No', 1: 'Yes'}

# Define category labels and calculate counts and rates
categories = (
    [f"Holiday Yesterday: {bool_map[val]}" for val in [0, 1]] +
    [f"Holiday Tomorrow:  {bool_map[val]}" for val in [0, 1]]
)
scheduled = []
no_shows = []
for col in ['was_holiday_yesterday', 'is_holiday_tomorrow']:
    for val in [0, 1]:
        subset = temp[temp[col] == val]
        tot = len(subset)
        ns_count = subset['no_show'].sum()
        scheduled.append(tot)
        no_shows.append(ns_count)

no_show_rate = [
    (ns / tot * 100) if tot > 0 else 0
    for ns, tot in zip(no_shows, scheduled)
]

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plot bars for scheduled and no-show
ax1.bar(categories, scheduled, alpha=0.5,
        label="Scheduled", edgecolor="black", color="gray")
ax1.bar(categories, no_shows, alpha=0.7,
        label="No-show", edgecolor="black", color="black")

# Annotate each bar with "scheduled/no-show"
for cat, tot, ns in zip(categories, scheduled, no_shows):
    ax1.annotate(f"{tot}/{int(ns)}", xy=(cat, tot),
                 ha="center", va="bottom", fontsize=8)

# Format primary axis
ax1.set_ylabel("Number of Instances")
plt.xticks(rotation=45)

# Create secondary axis for no-show rate
ax2 = ax1.twinx()
ax2.plot(categories, no_show_rate,
         marker="o", linestyle=":", label="No-show (%)", color="black")

# Set log scale and configure secondary-axis ticks
min_pos = min(r for r in no_show_rate if r > 0)
ax2.set_yscale("log")
ax2.yaxis.set_major_locator(LogLocator(
    base=10.0, subs=range(1, 10), numticks=15))
ax2.yaxis.set_major_formatter(FuncFormatter(
    lambda y, _: f"{y:.0f}%"))
ax2.set_ylim(min_pos * 0.8, max(no_show_rate) * 1.1)

# Format secondary axis
ax2.set_ylabel("No-show (%)")
ax2.tick_params(axis="y", labelsize=8)

# Combine legends and finalize
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2,
           loc="upper right")
ax1.spines["top"].set_visible(False)
ax2.spines["top"].set_visible(False)
plt.tight_layout()
plt.savefig("by_holiday.png", dpi=400)
plt.show()
```
